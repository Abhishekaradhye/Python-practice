```python
"""Next problem : 
Consider a string, S, that is a series of characters, each followed by its frequency as an integer. 
The string is not compressed correctly, so there may be multiple occurrences of the same character. 
A properly compressed string will consist of one instance of each character in alphabetical order 
followed by the total count of that character within the string.

Input : a3b2c5a1g9c2
Output : a4b2c7g9"""
```

```python
def CompressedStr(s):
    freq = {}
    i = 0 

    while i < len(s):
        char = s[i]
        i += 1

        number = ""

        while i < len(s) and s[i].isdigit():
            number += s[i]
            i += 1

        count = int(number)
        if char in freq:
            freq[char] += count
        else:
            freq[char] = count

    result = ""
    for ss in sorted(freq.keys()):
        result += ss +str(freq[ss])

    return result


CompressedStr('a3b2c5a1g9c2')
```

```python
"""Next Question : 
Given a strings that consists of lowercase English letters, select exactly one non-empty substring of s 
and replace each character of it with the previous character of the English alphabet. 
For example 'b' is converted to 'a', 'c' is converted to 'b',.... and 'a' is converted to 'Z'
Find the lexicographically smallest string that can be obtained after performing the above operation exactly once.

Example s="hackerrank"
Select New string
h gackerrank
ha gzckerrank
err hackdqqank"""
```

```python
def small_string(s):
    s = list(s)
    changed = False

    for i in range(len(s)):
        if s[i] != 'a':
            while i < len(s) and s[i] != 'a':
                s[i] = chr(ord(s[i]) - 1)

                i += 1
            changed = True
            break
    
    if not changed:
        s[-1] = 'z'

    return "".join(s)

print(small_string("hackerrank"))
print(small_string("abhishek"))
print(small_string("Zensar"))
print(small_string("Perfect"))
```

```python
"""Group Timestamps by Week
Problem Statement:
Given a list of timestamps in sequential order, group them into lists where each sublist contains timestamps from the same week, starting from the first timestamp.

Given Input:
["2023-01-01", "2023-01-02", "2023-01-08", "2023-01-09"]

Expected Output:
[["2023-01-01", "2023-01-02"], ["2023-01-08", "2023-01-09"]]"""
```

```python
from datetime import  datetime

def group_by_week(dates):
    dates_objs = [datetime.strptime(d, "%Y-%m-%d").date() for d in dates]
    start = dates_objs[0]

    week_map = {}

    for d in dates_objs:
        week_num = (d- start).days // 7

        if week_num not in week_map:
            week_map[week_num] = []
        
        week_map[week_num].append(d.strftime("%Y-%m-%d"))

    return [week_map[k] for k in sorted(week_map.keys())]


print(group_by_week(["2023-01-01", "2023-01-02", "2023-01-08", "2023-01-09"]))
```

```python
from urllib import request, parse
import json
from pprint import pprint
from datetime import datetime, timezone

endpoint = "https://earthquake.usgs.gov/fdsnws/event/1/query"

chars = {
    "format": "geojson",
    "starttime": "2024-01-01",
    "endtime": "2024-01-10",
    "minlatitude": 6.0,
    "maxlatitude": 36.0,
    "minlongitude": 68.0,
    "maxlongitude": 97.5
}


query = urllib.parse.urlencode(chars)

link = f"{endpoint}?{query}"

with urllib.request.urlopen(link) as eresponse:
    data = json.loads(eresponse.read().decode("utf-8"))


# Extract features and sort by time
events = data.get("features", [])
events.sort(key=lambda x: x["properties"]["time"])

pprint(events)

# Print sorted events with readable timestamp
for event in events:
    prop = event["properties"]
    timestamp_ms = prop["time"]
    dt = datetime.fromtimestamp(timestamp_ms / 1000, tz=timezone.utc)
    time_str = dt.strftime("%Y-%m-%d %H:%M:%S UTC")
    place = prop.get("place", "Unknown location")
    mag = prop.get("mag", "N/A")
    print(f"{time_str} | Magnitude: {mag} | Location: {place}")
```

```python
from urllib import request
import json
from pprint import pprint
from datetime import datetime

url_todo = "https://disease.sh/v3/covid-19/historical/India?lastdays=100"

headers = {
    "User-Agent": "Mozilla/5.0"
}

req = urllib.request.Request(url_todo, headers=headers)

with urllib.request.urlopen(req) as get_request:
    got_data = json.loads(get_request.read().decode())

timeline = got_data['timeline']

start_date = datetime(2022, 12, 1)
end_date = datetime(2023, 3, 5)

def filtered_timeline(timeline_section):
    filtered = {}
    for date_str, count in timeline_section.items():
        try:
            date_obj = datetime.strptime(date_str, "%m/%d/%y")
            if start_date <= date_obj <= end_date:
                filtered[date_obj]= count

        except:
            continue
    return dict(sorted(filtered.items()))

filtered_cases = filtered_timeline(timeline['cases'])
# print(filtered_cases)

print('\nCases')
for date,count in filtered_cases.items():
    print(f"{date.strftime('%Y-%m-%d')} : {count}")
```

```python
def isprime(num):
    if num <= 2 :
        return False
    for i in range(2,int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print(isprime(3))
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=' ')
        a, b = b, a + b

fibonacci(10)
```

```python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)
factorial(5)
```

```python
def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

factorial(6)
```

```python
def factorialnum(num):
    sol = 1
    for l in range(2,num+1):
        sol *=l
    return sol

factorialnum(5)
```

```python
def two_sum(nums, target):
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]

# Example usage
print(two_sum([2, 7, 5, 4, 15], 9))  # Output: [0, 1]
```
