```python
"""üîπ Problem 1 ‚Äî LISTS (manual logic, no built-ins for solving)

Given:
scores = [45, 67, 89, 45, 67, 90, 89, 45]

You must:

Remove duplicates manually (no set()).

Sort them in descending order manually (no sorted()).

Print highest and lowest clearly.

Concepts: loops, compare, swap, dedupe logic."""


scores = [45, 67, 89, 45, 67, 90, 89, 45]

the_scores = []

for i in scores:
    if i in the_scores:
        pass
    else:
        the_scores.append(i)

print(the_scores)

for s in range(len(the_scores)):
    for t in range(s+1, len(the_scores)):
        if the_scores[t] > the_scores[s]:
            the_scores[s], the_scores[t] = the_scores[t], the_scores[s]

print(the_scores)

print(f"The highest element is {the_scores[0]} and the lowest element is {the_scores[-1]}")
```

```python
"""Problem 2:
You are given a list of daily temperatures for a city:
temps = [30, 32, 29, 35, 40, 38, 28, 33]

You must do three tasks:
2.a ‚Üí Find the second highest temperature without sorting.
2.b ‚Üí Count how many days the temperature increased compared to the previous day.
2.c ‚Üí Create a new list that contains the difference between each consecutive day's temperatures (like ŒîTemp list)."""

temps = [30, 32, 29, 35, 40, 38, 28, 33]
highest = float("-inf")
higher = float("-inf")

for t in temps:
    if t > highest:
        higher = highest
        highest = t

    elif t > higher and t != highest:
        higher = t

print("Second highest temprature is : ",higher,".")

count = 0 
temp_diffs = []
for i in range(1,len(temps)):
    if temps[i] > temps[i-1]:
        count += 1
    temp_diffs.append(temps[i] - temps[i-1])
    
print(f"The temperature increased compared to the previous day {count} times.")

print(f"The differences between each consecutive day temperatures are {', '.join(map(str, temp_diffs[:-1]))} and {temp_diffs[-1]}.")
```

```python
# Exact age from date of birth

from datetime import date

dob_input = input("Enter DOB (YYYY-MM-DD): ")
y, m, d = map(int, dob_input.split('-'))
dob = date(y, m, d)
today = date.today()

# Calculate raw differences
years = today.year - dob.year
months = today.month - dob.month
days = today.day - dob.day

# Adjust if needed
if days < 0:
    months -= 1
    days += (date(today.year, today.month, 1) - date(today.year, today.month-1, 1)).days

if months < 0:
    years -= 1
    months += 12

print(f"Age: {years} years, {months} months, {days} days")
```

```python
# Factorial of a Number

def factorial(nmbr):
    if not isinstance(nmbr, int):
        raise ValueError("Number must be an integer.")
    
    elif nmbr < 0:
        raise Exception("Factorial is not defined for negative numbers.")
    
    elif nmbr == 0 or nmbr == 1:
        return 1
    
    count = 1
    for i in range(1,nmbr+1):
        count *= i
    return count

try:
    nmbr = int(input("Enter Number : "))
    print(f"Factorial of {nmbr} is {factorial(nmbr)}.")
except Exception as e:
    print(e)
```

```python
# Problem:
# Given an integer array, return the length of the longest subarray where the "element" is strictly increasing after every element.
# Example:
# Input: [1, 3, 2, 5, 9]
# Output: 3

def long_len(lst):
    diffs = []
    diff = 0
    for i in range(1,len(lst)):
        if lst[i] - lst[i-1] > 0:
            # print(f"{lst[i]} - {lst[i-1]} ")
            diff += 1

        elif lst[i] - lst[i-1] <= 0:
            diff = 0
        
        diffs.append(diff+1 if diff > 0 else 0)

    return max(diffs) if diffs else len(lst)

print(long_len([1, 3, 2, 5, 9]))
    
```

```python
# Problem:
# Given an integer array, return the length of the longest subarray where the "Sum" is strictly increasing after every element.
# Example:
# Input: [1, 3, 2, 5, 9]
# Valid longest: [2 ‚Üí sum(2)=2] , [2,5 ‚Üí 7] , [2,5,9 ‚Üí16] (increasing)
# Output: 3

def longest_increasing_subarray(arr):
    if not arr:
        return 0
    
    max_len = 1
    current_len = 1
    current_sum = arr[0]
    
    for i in range(1, len(arr)):
        if current_sum + arr[i] > current_sum:
            current_len += 1
            current_sum += arr[i]
        else:
            current_len = 1
            current_sum = arr[i]
        max_len = max(max_len, current_len)
    
    return max_len

tests = [[1], [5, 4, 3], [1, 2, 3], [3, 1, 2, 4], [10, 10, 10], [1, -1, 2, -2, 3], [0, 0, 1, 2], [-5, -4, -3, -2], [2, 2, 3, 3, 4], [100, 1, 2, 3, 4, 5], [1, 100, -50, 200], [1, 3, 5, -2, 7, 9], [2, -1, -2, -3, 10], [-1, -2, -3, -4], [5, -1, 5, -1, 5], [3, 3, -1, 3, 3], [1, 2, -3, 4, 5, -2, 6], [50, -49, -1, 100], [1, 2, 1, 2, 1, 2], [0, -1, 1, -1, 2, -1, 3]]



anssss = list(map(longest_increasing_subarray, tests))
print(anssss)
```

```python
"""‚úÖ TASK 2 ‚Äî Mini OOP Problem (Real-life Inspired)

Problem:
Create a ParkingMeter class:

User adds money

Each ‚Çπ10 buys 15 minutes

Meter should track: balance, minutes purchased, and minutes remaining

use(minutes) should deduct time and auto-stop if usage exceeds balance

status() shows remaining minutes and balance

Think like a real parking meter ‚Üí controlled, safe, edge-case clean"""


class ParkingMeter:
    def __init__(self):
        self.balance = 0
        self.minutes_remaining = 0
    
    def add_money(self, amount):
        if amount <= 0:
            return
        
        usable_amount = (amount // 10) * 10
        self.balance += amount - usable_amount

        minutes_added = (usable_amount // 10) * 15
        self.minutes_remaining += minutes_added

    def use(self, minutes):
        if minutes <= 0 or self.minutes_remaining == 0:
            return
        
        elif minutes >= self.minutes_remaining:
            self.minutes_remaining = 0

        else:
            self.minutes_remaining -= minutes

    def status(self):
        return{"Minutes Remaining":  self.minutes_remaining, "\nAccount Balance":  self.balance}


parkingmeter = ParkingMeter()

parkingmeter.add_money(557)
print(parkingmeter.status())
parkingmeter.use(300)
print(parkingmeter.status())
```
