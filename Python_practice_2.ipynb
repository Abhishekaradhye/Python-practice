```python
"""Loop + Accumulation
Take an integer input n.
Print the sum of all even numbers from 1 to n.
(Concept: for loop, modulo operator %, running total)"""

n = 50

s = 0

for i in range(1,n+1):
    if i % 2 == 0:
        s += i

print(f"Sum of all even numbers from 1 to {n} is {s}.")
```

```python
"""List + Basic Algorithm
Given a list of 5 integers entered by the user, print the maximum and minimum numbers manually (no max() or min() built-ins).
(Concept: iteration, condition checking, variable tracking)"""

nums = [int(x) for x in input("Enter 5 numbers separated by commas: ").split(',')]
mx = nums[0]
mn = nums[0]
for n in nums:
    if n > mx: mx = n
    if n < mn: mn = n
print(f"Max = {mx}, Min = {mn}")
```

```python
"""üîπ Problem 1 ‚Äî LISTS (manual logic, no built-ins for solving)

Given:
scores = [45, 67, 89, 45, 67, 90, 89, 45]

You must:

Remove duplicates manually (no set()).

Sort them in descending order manually (no sorted()).

Print highest and lowest clearly.

Concepts: loops, compare, swap, dedupe logic."""


scores = [45, 67, 89, 45, 67, 90, 89, 45]

the_scores = []

for i in scores:
    if i in the_scores:
        pass
    else:
        the_scores.append(i)

print(the_scores)

for s in range(len(the_scores)):
    for t in range(s+1, len(the_scores)):
        if the_scores[t] > the_scores[s]:
            the_scores[s], the_scores[t] = the_scores[t], the_scores[s]

print(the_scores)

print(f"The highest element is {the_scores[0]} and the lowest element is {the_scores[-1]}")
```

```python
"""Problem 2:
You are given a list of daily temperatures for a city:
temps = [30, 32, 29, 35, 40, 38, 28, 33]

You must do three tasks:
2.a ‚Üí Find the second highest temperature without sorting.
2.b ‚Üí Count how many days the temperature increased compared to the previous day.
2.c ‚Üí Create a new list that contains the difference between each consecutive day's temperatures (like ŒîTemp list)."""

temps = [30, 32, 29, 35, 40, 38, 28, 33]
highest = float("-inf")
higher = float("-inf")

for t in temps:
    if t > highest:
        higher = highest
        highest = t

    elif t > higher and t != highest:
        higher = t

print("Second highest temprature is : ",higher,".")

count = 0 
temp_diffs = []
for i in range(1,len(temps)):
    if temps[i] > temps[i-1]:
        count += 1
    temp_diffs.append(temps[i] - temps[i-1])
    
print(f"The temperature increased compared to the previous day {count} times.")

print(f"The differences between each consecutive day temperatures are {', '.join(map(str, temp_diffs[:-1]))} and {temp_diffs[-1]}.")
```

```python
"""Problem 3 ‚Äî ‚ÄúEmployee Salary Analyzer‚Äù
(Intermediate - clean loops, comparisons, simple math)

You are given a list of employee salaries (monthly, in ‚Çπ):
salaries = [28000, 45000, 32000, 45000, 60000, 29000, 60000, 31000]

You must do three tasks:

3.a ‚Üí Find the two lowest distinct salaries.
(No sorting allowed. No built-ins like min().)

3.b ‚Üí Count how many employees earn more than ‚Çπ40,000.
(Straightforward loop.)

3.c ‚Üí Create a new list where each value is the salary + 10% bonus.
(Bonus = salary * 0.10)"""

salaries = [28000, 45000, 32000, 45000, 60000, 29000, 60000, 31000]

one = float("inf")
two = float("inf")

for s in salaries:
    if s < one:
        two = one
        one = s

    elif s < two and s != one:
        two = s

print("the two lowest distinct salaries are : ",one,",", two)

count = 0

count = sum( s > 40000 for s in salaries)
print(count, ' employees earn more than ‚Çπ40,000.')

bonus = [f"{int(s+(s*.1)):,}" for s in salaries]

print(bonus)
```
